```python
from pathlib import Path
from pprint import pp
import json

import pandas as pd

itell_to_prolific_id_map = (
    pd.read_csv(
        "/home/jovyan/active-projects/2024.05-CTTC/data/supabase_filtered/users.csv"
    )
    .set_index("id")["prolific_pid"]
    .to_dict()
)

users_to_keep = (
    pd.read_csv(
        "/home/jovyan/active-projects/2024.05-CTTC/data/itell_cttc_final_ids.csv"
    )["itell_id"]
)

intake = (
    pd.read_csv(
        "/home/jovyan/active-projects/2024.05-CTTC/data/surveys/iTELL Intake - CTTC_August 24, 2024_16.42.csv"
    )
    .query("not PQ15.isnull()") # Only responses that have a value for the last question
)

df = (
    pd.read_csv(
        "/home/jovyan/active-projects/2024.05-CTTC/data/supabase_filtered/constructed_responses.csv"
    )
    .query("user_id in @users_to_keep")
    .rename(columns={"text":"response"})
    .assign(
        prolific_pid = lambda x: x.user_id.map(itell_to_prolific_id_map)
    )
    .merge(intake,
           left_on="prolific_pid",
           right_on="PID",
           how="left",
          )
    .set_index("created_at")
    .rename_axis(None) # Remove index name
    .sort_index()
    .drop_duplicates(subset=["user_id", "chunk_slug"], keep="first")
)
    
df
```

```python
content_dir = Path("/home/jovyan/active-projects/shared-resources/old_texts_from_strapi_json/")

content_list = []
for fn in content_dir.iterdir():
    with open(fn) as f:
        volume = json.load(f)["data"]["attributes"]
        for page_data in volume["Pages"]["data"]:
            page = page_data["attributes"]
            for chunk in page["Content"]:
                d = {}
                if not chunk["__component"] == "page.chunk":
                    continue
                d["volume_title"] = volume["Title"]
                d["volume_slug"] = volume["Slug"]
                d["page_title"] = page["Title"]
                d["page_slug"] = page["Slug"]
                d["chunk_header"] = chunk["Header"]
                d["chunk_slug"] = chunk["Slug"]
                d["chunk_text"] = chunk["CleanText"]
                d["question"] = chunk["Question"]
                d["answer"] = chunk["ConstructedResponse"]
                content_list.append(d)

content = pd.DataFrame(content_list).set_index("chunk_slug")
content
```

```python
df.to_csv("data/cri_responses.csv")
content.to_csv("data/cri_content.csv")
```
