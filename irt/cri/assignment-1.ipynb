```python
import pandas as pd

binary_score_map = {
    0:0,
    1:1,
    2:1,
}

df = pd.read_csv("data/cri_responses.csv", index_col=0)
df["binary_score"] = df.score.map(binary_score_map)
content = pd.read_csv("data/cri_content.csv", index_col=0)
df
```

```python
print(f"Num Users: {df.prolific_pid.nunique()}")
print(f"Num Items: {df.chunk_slug.nunique()}")
items_answered = df.groupby("prolific_pid").chunk_slug.nunique()
print(f"Items Answered Mean: {items_answered.mean():.3f}")
print(f"Items Answered Min: {items_answered.min()}")
print(f"Items Answered Max: {items_answered.max()}")
```

```python
from IPython.display import HTML

crosstab = (
    pd.crosstab(df.chunk_slug, df.binary_score, normalize="index")
    .rename(columns={
        0: "Incorrect",
        1: "Correct",
    })
)
crosstab["Difficulty"] = crosstab["Correct"]

# Calculate total score for each student
df["total_score"] = df.groupby("prolific_pid")["binary_score"].transform("sum")
df["mean_score"] = df.groupby("prolific_pid")["binary_score"].transform("mean")

# Calculate item discrimination for each item
crosstab["Discrimination (Total)"] = df.groupby("chunk_slug").apply(
    lambda x: x["binary_score"].corr(x["total_score"])
)
crosstab["Discrimination (Mean)"] = df.groupby("chunk_slug").apply(
    lambda x: x["binary_score"].corr(x["mean_score"])
)
output = crosstab.reset_index().reset_index().drop(columns="chunk_slug").rename_axis(None, axis=1).round(3)
display(HTML(output.to_html(index=False)))
```
